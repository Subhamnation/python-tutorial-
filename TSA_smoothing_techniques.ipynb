### Moving Average 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample time series data
np.random.seed(0)
data = np.random.normal(0, 1, 100).cumsum()  # cumulative sum to simulate trend
df = pd.DataFrame(data, columns=['Value'])

# Moving Average Smoothing
window = 5
df['Moving_Avg'] = df['Value'].rolling(window=window).mean()


# Plotting
plt.plot(df['Value'], label='Original')
plt.plot(df['Moving_Avg'], label=f'Moving Average (window={window})', color='red')
plt.legend()
plt.show()

```

### Exponential Smoothing 


```python

# Exponential Smoothing
alpha = 0.3  # smoothing factor
df['Exp_Smoothing'] = df['Value'].ewm(alpha=alpha).mean()

# Plotting
plt.plot(df['Value'], label='Original')
plt.plot(df['Exp_Smoothing'], label=f'Exponential Smoothing (alpha={alpha})', color='green')
plt.legend()
plt.show()
```

### Gaussian Smoothing

```python
from scipy.ndimage import gaussian_filter1d

# Gaussian Smoothing
sigma = 2
df['Gaussian_Smoothing'] = gaussian_filter1d(df['Value'], sigma=sigma)

# Plotting
plt.plot(df['Value'], label='Original')
plt.plot(df['Gaussian_Smoothing'], label=f'Gaussian Smoothing (sigma={sigma})', color='purple')
plt.legend()
plt.show()
```

### Savitzky-Golay Filter 

```python
from scipy.signal import savgol_filter

# Savitzky-Golay Smoothing
window_length = 7
poly_order = 2
df['Savitzky_Golay'] = savgol_filter(df['Value'], window_length=window_length, polyorder=poly_order)

# Plotting
plt.plot(df['Value'], label='Original')
plt.plot(df['Savitzky_Golay'], label=f'Savitzky-Golay (window={window_length}, polyorder={poly_order})', color='orange')
plt.legend()
plt.show()
```
